***Bartender - Agente Reactivo Simple***

Primero hay que ejecutar los bloques 2 y 3, ya luego en el 4to se hacen los ingresos.

```python
reglasBartender = {
    "hola": "Bienvenido al bar, ¿qué te gustaría tomar? Puedo darte el menu si lo deseas",
    "menu": "Tenemos: mojito, margarita, piña colada, daiquiri, manhattan, negroni, cerveza, vino.",
    "mojito": "Preparando Mojito: Ron, menta, limón, azúcar y... soda. ¡Listo!",
    "margarita": "Preparando Margarita: Tequila, triple sec, limón y... sal. ¡Salud!",
    "piña colada": "Preparando Piña Colada: Ron, crema de coco y... piña. ¡Tropical!",
    "pinacolada": "Preparando Piña Colada: Ron, crema de coco y... piña. ¡Muy tropical!",
    "daiquiri": "Preparando Daiquiri: Ron blanco, limón y azúcar. ¡Bastante refrescante!",
    "manhattan": "Preparando Manhattan: Whisky, vermut y bitter. ¡Un clásico!",
    "negroni": "Preparando Negroni: Gin, campari y vermut. ¡El amargo perfecto!",
    "cerveza": "Sirviendo una cerveza bien fría. ¡Disfrutala!",
    "vino": "Sirviendo una copa de vino. ¿Tinto o blanco?",
    "tinto": "Excelente elección, vino tinto servido",
    "blanco": "Perfecto, vino blanco bien frío",
    "dulce": "Te recomiendo: Piña Colada o Daiquiri",
    "fuerte": "Te recomiendo: Manhattan o Negroni",
    "refrescante": "Te recomiendo: Mojito o Margarita",
    "ron": "Con ron hay: Mojito, Piña Colada o Daiquiri",
    "tequila": "Con tequila hay: La margarita",
    "whisky": "Con whisky hay: Manhattan",
    "gin": "Con gin hay: Negroni",
    "sorpresa": "¡Sorpresa del bartender! Te preparo un Mojito especial",
    "gracias": "¡De nada! Vuelve cuando quieras",
    "adios": "¡Hasta pronto! Bebe con responsabilidad"
}
```

```python
class AgenteBartender:
  def __init__(self, reglasBartender):
    self.reglasBartender = reglasBartender

  def actuar(self, percepcion):
      percepcion = percepcion.lower()
      if percepcion in self.reglasBartender:
          accion = self.reglasBartender[percepcion]
          return accion
      else:
          return "No tengo esa bebida. Vuelve a leer el menu porfavor."
```

```python
bartender = AgenteBartender(reglasBartender)
percepcion = input("Percepción -> ")
while percepcion != "salir":
  print(bartender.actuar(percepcion))
  percepcion = input("Percepción -> ")
print("¡Gracias por visitar el bar! Hasta luego")
```
